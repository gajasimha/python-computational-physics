# Plots

```python
import matplotlib.pyplot as plt
import numpy as np
```

Plot

```python
coords1 = np.array([[1, 1, 1], [4, 5, 6], [7, 8, 9]])
```

```python
x1, y1, z1 = coords1.T
plt.xlabel('x')
plt.ylabel('y, z')
plt.scatter(x1, y1)
plt.scatter(x1, z1)
```

```python
x1 = np.linspace(0, 4, 100) # (start, end, number of points)
y1 = 1 + 2 * np.cos(3*x1)
plt.scatter(x1, y1)
plt.xlim(0, 4)
```

Density Plot

```python
x2, y2 = np.meshgrid(
    np.linspace(-10, 10, 201),
    np.linspace(-10, 10, 201)
)
z2 = x2 + y2**2

ct = plt.contour(x2, y2, z2)
plt.clabel(ct, inline = 1,  fontsize=10)
```

```python
x3, y3 = np.meshgrid(
    np.linspace(-10, 10, 201),
    np.linspace(-10, 10, 201)
)
z3 = np.cos(x3+y3) + 0.05*(x3-y3)

plt.contourf(x3, y3, z3, levels = 20)
plt.colorbar()
```

3D Plots

```python
plotproj = plt.axes(projection='3d')
plotproj.contour3D(x3, y3, z3, 100, cmap='binary')
```

```python
plt3D = plt.axes(projection='3d')
z1 = np.linspace(0, 30, 301)
x1 = np.sin(z1)
y1 = np.cos(z1)

plt3D.scatter3D(x1, y1, z1)
```
